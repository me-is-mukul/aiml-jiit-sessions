```python
count = 0
while count < 4:
    print(count)
    count += 1   # there are no ++ and -- operators in python
```

```python
count = 0
while count < 4:
    print(count)
    count += 1

else:
    print("Loop finished without hitting a break") #else in while loop in python triggers if the while loops completes successfully
```

Data Structures

* List: [1,2,3,"a","b","c"]     Mutable, Ordered, Sliceable
* Tuple: (1, 2, 3, a","b")      Immutable, Ordered, Sliceable
* Dictionary {"key": "value"}   Mutable, Ordered (Insertion Order), Not Sliceable
* Sets {1,2,3,"A","B"}          Mutable, Unordered, Not Sliceable

```python
mdict = {3:'apple', 1:'banana', 2:'cherry'}

tup1=("A","B")   #inserting an immutable datatype as key
mdict[tup1]='pie' #values can be of any datatype

for fruit in mdict:
  print(fruit, mdict[fruit])
```

```python
mdict = {True:'apple', 1:'banana', 1.0:'cherry'}

for fruit in mdict:
  print(fruit, mdict[fruit])
```

```python
my_set = {'apple', 'banana', 'cherry'} #only takes immutable datatypes as elements

for food in my_set:
    print(food)
```

```python
my_set.add("pie")

for food in my_set:
    print(food)
```

FROZENSET : Read-only version of Set (Immutable)

```python
frznset=frozenset(my_set)

#doing frznset.add("Item") will fail

for food in my_set:
    print(food)
```
