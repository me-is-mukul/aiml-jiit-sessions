## solving system of linear equations

```python
A = np.array([
    [2,1,-1],
    [-3,-1,2],
    [-2,1,2]
])

B = np.array([8,-11,-3])

solution = np.linalg.solve(A,B)

a = np.array([1,2,3])
b = np.array([[3],[2],[1]])
a @ b
```

## roation of a coordinate using numpy

```python
p = np.array([3,2])
```

```python
theta = np.radians(45)
print(type(theta))
```

```python
rotation = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta), np.cos(theta)]
])
```

```python
new_coordinate = rotation @ p
```
