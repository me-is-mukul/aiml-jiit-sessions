## Data cleaning

1. Handling missing values

```python
## Check missing values

df.isna().sum()
```

Common ways of filling missing values

-> Fill with mean / median (for numeric columns) (Median preferred when skewed data/outliers)

-> Fill with mode (for categorical data)

-> Fill with 0

-> Remove entire row/column

```python
## Fill age with mean

df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["Age"].isna().sum()
```

```python
df["TypeofContact"] = df["TypeofContact"].fillna(df["TypeofContact"].mode()[0])
df["TypeofContact"].isna().sum()
```

## ENCODING => Converting categorical column to numerical column
### Types:
    1. Label Encoding (Ordinal)
    2. One hot Encoding (Nominal)

## LABEL ENCODING

Cup size: 

small-> 0

medium -> 1

large -> 2


2 > 1 > 0

large > medium > small



```python
df = pd.DataFrame({
    "Customer": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Cup Size": ["Small", "Medium", "Large", "Small", "Large"]
})

df
```

```python
size_mapping = {
    "Small": 0,
    "Medium": 1,
    "Large": 2
}

df["Cup Size"] = df["Cup Size"].map(size_mapping)

df
```
