### Series : 1-d array like object, can hold any datatype, similar to column in a table

```python
## creating series from a list

data=[1,2,3,4,5]
series = pd.Series(data)
print("series \n" ,series)
print(type(series))
```

```python
## create series from a dictionary

data={'a':1, 'b':2, 'c':3}

series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data, index=index)
```

### DataFrame : 2-dimensional data structure with labeled axes (rows and columns).

#### Ways of creating DataFrame

1. Create a dataframe from a dictionary of a list
2. create a dataframe from a list of dictionaries
3. Loading data from csv

```python
## create a dataframe from dictionary of list

data = {
    'name':['Krish', 'John', 'Jack'],
    'age' : [25,30,45],
    'city' : ['Bangalore', 'New York', 'Florida']
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
## create a dataframe from a list of dictionaries

data = [
    {'name':'Krish', 'Age':32, 'City': 'Bangalore'},
    {'name': 'John', 'Age':34, 'City':'Bangalore'},
    {'name':'Bappy', 'Age':32, 'City':"Bangalore"},
    {'name':'Jack', 'Age':32, 'City': 'Bangalore'}
]

df = pd.DataFrame(data)
print(df)
print(type(df))
```
